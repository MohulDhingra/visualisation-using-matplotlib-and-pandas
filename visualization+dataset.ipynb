```python
import pandas as  pd
import matplotlib.pyplot as mp


data= pd.read_csv("sample_book_data.csv")

mp.bar( data['Book Name'],data['Rating'])
mp.show()
```

```python
import pandas as  pd
import matplotlib.pyplot as mp


data= pd.read_csv("sample_book_data.csv")
mp.figure(figsize=(80, 80))

mp.plot(data['Rating'], data['Price'])
mp.show()
```

```python
import pandas as  pd
import matplotlib.pyplot as mp


data= pd.read_csv("sample_book_data.csv")
mp.figure(figsize=(80, 80))

data=data.sort_values('Price')

mp.plot(data['Rating'], data['Price'])
mp.show()
```
